# Storing data into dictionary of dfs

```python
# create an empty dictionary
dfs = {}

# opens a connection to the databse
with duckdb.connect("data/sakila.duckdb") as conn: 
    # loop through each table name from the 'description' data frame
    for name in description["name"]:
    # skip all tables whose name starts with an underscore (ex. _dlt_loads, _dlt_version)
       # name[0] = the first character of the table name
        # != '_' = "is NOT an underscore"
        if name[0] != '_':

            # Load Each Table as a Dataframe
        
            dfs[name] = conn.sql(f"FROM staging.{name};").df()

            # f"FROM staging.{name}; --> Creates a SQL query using the table name.
            # ex. name = "actor", this becomes FROM staging.actor;

            # conn.sql(...) = Run the SQL query
            # .df() = Convert the results into a pandas dataframe
            
dfs["actor"].head()
```
